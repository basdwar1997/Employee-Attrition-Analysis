```python
percentage_missing_values =df.isnull().sum()*100/len(df)
percentage_missing_values
```

```python
delet_columns=percentage_missing_values[percentage_missing_values>70].index
len(delet_columns)
df_clean=df.drop(columns=delet_columns)
```

```python
df_clean =df_clean.dropna(subset=['name','normalized_name'],axis=0)
df_clean.head(10)
```

```python
df_clean['domain']=df_clean['domain'].fillna('unknown')
df_clean['overview']=df_clean['overview'].fillna('unknown')
df_clean['logo_url']=df_clean['logo_url'].fillna('0')
df_clean['homepage_url']=df_clean['homepage_url'].fillna('unknown')
df_clean['founded_at']=df_clean['founded_at'].fillna('unknown')
mean_value=df_clean['relationships'].mean()
df_clean['relationships']=df_clean['relationships'].fillna(mean_value)
df_clean['created_by']=df_clean['created_by'].fillna('unknown')
df_clean['updated_at']=df_clean['updated_at'].fillna('unknown')
```

```python
percentage_missing_values =df_clean.isnull().sum()*100/len(df)
percentage_missing_values
```

```python
# all the categorical variables
for i in df_clean.columns:
  if df_clean[i].dtype=='object':
    print(i,df_clean[i].nunique())
```

```python
#now label encoder
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder() # Use LabelEncoder instead of label_encoder
df_clean['country_code']=le.fit_transform(df_clean['country_code'])
```
